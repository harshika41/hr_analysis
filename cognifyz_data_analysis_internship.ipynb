```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#level 1
#Task 1 A.(Top 3 Cuisines )
Cuisines_3=df['Cuisines'].str.split(',').explode().value_counts()

#Top 3 Cuisines from Database
Top_3=Cuisines_3.head(3)
print("Top 3 Cuisines From Dataset:")
print(Top_3)
```

```python
#Task 1 B.(Percentage of restaurants selling these cuisines)
restaurants= df.shape[0]
Percentage_of_restaurants = (Top_3/restaurants)*100
print(Percentage_of_restaurants)
```

```python
#LEVEL 1
#Task 2A: Cities with highest no. of restaurants
cities= df['City'].value_counts()
restaurants= cities.idxmax()
number_of_restaurants= cities.max()
print("City with highest restaurants", restaurants)
print("Number of restaurants", number_of_restaurants)
```

```python
#Task 2B: Average_Rating_for_restaurants_in_each_city
average_rating= df.groupby('City')['Aggregate rating'].mean()
print(average_rating)
```

```python
#Task 2C: City_with_highest_rating
print(average_rating.idxmax())
```

```python
#Task 2C rating_of_highest_rated_city
print(average_rating.max())
```

```python
#Task 3A- Create a histogram of distribution of price ranges
sns.histplot(df, x='Price range', color='violet', kde= True)
plt.title('PRICE RANGE DISTRIBUTION')
plt.xlabel('Price Range', fontsize=14)
plt.ylabel('Restaurant count',fontsize=14)
plt.show()
```

```python
#Task 3B- percentage of restaurants in each price category
df['Price range'].value_counts(normalize=True)*100
```

```python
#Task 4A- restaurant offering service of online delivery
df['Has Online delivery'].value_counts()
```

```python
percentage_of_restaurants_with_online_delivery=((df['Has Online delivery'] == 'Yes').sum()/df['Has Online delivery'].count())*100
percentage_of_restaurants_with_online_delivery
```

```python
#Task 4B- Comparing average ratings of restaurants
sns.barplot(df, y='Aggregate rating', color='red', hue= 'Has Online delivery')
plt.title('AVERAGE RATING OF RESTAURANTS')
```

```python
#LEVEL2
#Task 1 A.(Distribution of aggregate ratings and the most common ratings)
rated_data = df[df['Aggregate rating']>0]
rated_data['Aggregate rating'].hist(bins=10,color='green',edgecolor='white', figsize=(10, 5))
plt.title('Distribution of Aggregate Ratings', fontsize=16)
plt.xlabel('Aggregate Rating',fontsize=14)
plt.ylabel('Restaurant Count',fontsize=14)
plt.show()
```

```python
#Most Common Rating
most_common_rating=rated_data['Aggregate rating'].mode()[0]
most_common_rating
```

```python
#Level 2
#Task 1 B.Average votes
df['Votes'] = pd.to_numeric(df['Votes'],errors='coerce')
average_votes = df['Votes'].mean()

average_votes
```

```python
#Level 2
#Task 2 A.Most Common Cuisines
common_cuisines=df['Cuisines'].mode()[0]
common_cuisines
```

```python
#Level 2
#Task 2 B.Cuisnes Combination to have higher ratings
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'],errors='coerce')
average_rating=df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
average_rating.head(10)
```

```python
#Level 2
#Task 3 A.(locations of restaurants on a map)
Latitude=df['Latitude']
Longitude=df['Longitude']

plt.figure(figsize=(10,6))
plt.scatter(Longitude,Latitude,color='yellow',edgecolor='black')
plt.title('LOCATIONS',fontsize=14)
plt.xlabel('Longitude',fontsize=14)
plt.ylabel('Latitude',fontsize=14)
plt.grid()
plt.show()
```

```python
#Level 2
#Task 4 A.( Restaurant chain from dataset)
df['Restaurant Name'].value_counts()
```

```python
#Level 2
#Task 4 B.(ratings and popularity of different restaurant chains)
restaurants_analysis = df.groupby('Restaurant Name').aggregate({'Aggregate rating':'mean','Votes':'sum','Price range':'mean'}).reset_index()
top_restaurants = restaurants_analysis.sort_values(by='Aggregate rating',ascending=False).head(10)
top_restaurants
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_restaurants,x='Aggregate rating',y='Restaurant Name')
plt.title('Top 10 restaurant with highest rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Restaurant Name')
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(data=top_restaurants,x='Votes',y='Restaurant Name',color='pink',edgecolor='blue')
plt.title('Top 10 restaurants by popularity (Votes)')
plt.xlabel('Votes')
plt.ylabel('Restaurant Name')
plt.show()
```

```python
price_analysis=df.groupby('Price range').aggregate({'Aggregate rating':'mean','Votes':'sum'}).reset_index()

plt.figure(figsize=(8,4))
sns.barplot(data=price_analysis,x='Price range',y='Aggregate rating',color='blue',edgecolor='yellow')
plt.title('Average Rating vs Price range')
plt.xlabel('Price range')
plt.ylabel('Average rating')
plt.show()
```
